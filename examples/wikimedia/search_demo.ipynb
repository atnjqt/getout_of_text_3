```python
# Install datasets if needed (uncomment to run)
# !pip install datasets

from examples.wikimedia.search_wikimedia import search_dataset, search_dataset_multithreaded

dataset_dir = './examples/wikimedia/wikimedia_en'  # adjust path if needed
keyword = 'bank'

# CLI example (run in shell):
# python examples/wikimedia/search_wikimedia.py --dataset-dir ./examples/wikimedia/wikimedia_en --keyword bank --max-results 5 --workers 4
```

```python
# Single-threaded example
hits = list(search_dataset(dataset_dir, keyword, max_results=3, context=20))
for i, h in enumerate(hits, 1):
    print(f'STD [{i}] index={h["index"]} split={h["split"]}')
    print(h['snippet'])
    print('-'*80)
```

```python
# Multithreaded / Multiprocess example (3 workers)
# The library will prefer multiprocessing when workers>1 on most platforms
hits_mt = list(search_dataset_multiprocessed(dataset_dir, keyword, max_results=3, context=20, workers=3))
for i, h in enumerate(hits_mt, 1):
    print(f'MP  [{i}] index={h["index"]} split={h["split"]}')
    print(h['snippet'])
    print('-'*80)
```
