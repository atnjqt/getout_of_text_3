````python
# GloWbe Computational Forensic Linguistics Agent
# Adapted from SCOTUS analysis tools for corpus linguistics analysis

from langchain.tools import BaseTool
from langchain.pydantic_v1 import BaseModel, Field
from langchain.chat_models import init_chat_model
from typing import Optional, Type, Dict, Any, Union, List
import json
import re
from datetime import datetime


class GloWbeAnalysisInput(BaseModel):
    """Input for GloWbe corpus linguistics analysis tool."""
    keyword: str = Field(description="The keyword/phrase to analyze from GloWbe KWIC results")
    results_json: Union[str, Dict[str, Any]] = Field(
        description="Pre-filtered GloWbe KWIC JSON results from got3.search_keyword_corpus"
    )
    analysis_focus: Optional[str] = Field(
        default="forensic_linguistics", 
        description="Analysis approach: 'forensic_linguistics', 'semantic_variation', 'register_analysis', 'diachronic', 'comparative'"
    )
    max_contexts: Optional[int] = Field(
        default=None, description="DEPRECATED: No longer used. Tool processes all provided contexts."
    )
    return_json: bool = Field(
        default=False, description="If True, return structured JSON with reasoning and findings"
    )
    extraction_strategy: str = Field(
        default="all",
        description="Text extraction: 'first', 'all', or 'raw_json'"
    )
    debug: bool = Field(default=False, description="Enable debug metrics")


class GloWbeForensicLinguisticsTool(BaseTool):
    """
    AI tool for computational forensic linguistics analysis of GloWbe KWIC results.

    Applies systematic data science, legal scholarship, and applied linguistics 
    methodologies to analyze keyword usage patterns across GloWbe genres.
    """
    name: str = "GloWbe_forensic_analysis"
    description: str = (
        "Performs computational forensic linguistics analysis on GloWbe KWIC results "
        "The corpus of Global Web-based English (GloWbE; pronounced `globe`) was created by Mark Davies, and it is unique in the way that it allows you to carry out comparisons between different varieties of English. GloWbE is related to other corpora from English-Corpora.org, which are the most widely used corpora of English and which offer unparalleled insight into variation in English."
        "GloWbE contains about 1.9 billion words of text from twenty different countries. This makes it about 100 times as large as other corpora like the International Corpus of English, and it allows for many types of searches that would not be possible otherwise. In addition to this online interface, you can also download full-text data from the corpus."
        "using data science and applied linguistics methodologies."
    )
    args_schema: Type[BaseModel] = GloWbeAnalysisInput
    model: Any = Field(exclude=True)

    def __init__(self, model, **kwargs):
        super().__init__(**kwargs)
        self.model = model

    def _run(
        self,
        keyword: str,
        results_json: Union[str, Dict[str, Any]],
        analysis_focus: str = "forensic_linguistics",
        max_contexts: Optional[int] = None,
        return_json: bool = False,
        extraction_strategy: str = "all",
        debug: bool = False,
    ) -> Union[str, Dict[str, Any]]:
        try:
            return self._execute(keyword, results_json, analysis_focus, max_contexts, return_json, extraction_strategy, debug)
        except Exception as e:
            error_str = str(e)
            return f"‚ùå Error during GloWbe forensic analysis: {error_str}"

    async def _arun(
        self,
        keyword: str,
        results_json: Union[str, Dict[str, Any]],
        analysis_focus: str = "forensic_linguistics",
        max_contexts: Optional[int] = None,
        return_json: bool = False,
        extraction_strategy: str = "all",
        debug: bool = False,
    ) -> Union[str, Dict[str, Any]]:
        return self._run(keyword, results_json, analysis_focus, max_contexts, return_json, extraction_strategy, debug)

    def _execute(self, keyword, results_json, analysis_focus, max_contexts, return_json, extraction_strategy, debug):
        # Parse and validate input
        results_dict = self._parse_GloWbe_results(results_json)
        stats = self._compute_GloWbe_stats(results_dict, keyword, extraction_strategy)
        
        # Extract contexts and estimate token usage
        contexts = self._extract_contexts(results_dict, max_contexts, extraction_strategy)
        
        # Debug metrics
        if debug:
            print("‚úÖ Reading GloWbe results for keyword:", keyword)
            raw_chars = len(json.dumps(results_dict))
            extracted_chars = sum(len(c) for c in contexts)
            print(f"üß™ GloWbe DEBUG: genre_year_keys={len(results_dict)} raw_chars={raw_chars} extracted_chars={extracted_chars} total_contexts={len(contexts)}")
            
            # Debug: Show genre distribution in ALL extracted contexts
            genre_context_counts = {}
            for context in contexts:
                if context.startswith('[') and ':' in context:
                    genre = context.split(':')[0][1:]  # Extract genre from [genre:year:filename]
                    genre_context_counts[genre] = genre_context_counts.get(genre, 0) + 1
            print(f"üéØ All extracted contexts by genre: {genre_context_counts}")
            print(f"üìä Total contexts extracted: {len(contexts)}")
        
        # Build specialized prompt and check token limits
        prompt = self._build_GloWbe_prompt(keyword, results_dict, stats, analysis_focus, max_contexts, return_json, extraction_strategy)
        # Invoke model
        response = self.model.invoke([{"role": "user", "content": prompt}])
        content = getattr(response, 'content', str(response))
        
        if return_json:
            return self._postprocess_GloWbe_json(content, stats)
        return content

    def _parse_GloWbe_results(self, results_json: Union[str, Dict[str, Any]]) -> Dict[str, Any]:
        """Parse GloWbe results JSON - expects genre->subkey->dataframe structure."""
        if isinstance(results_json, str):
            results_dict = json.loads(results_json)
        else:
            results_dict = results_json
        
        if not isinstance(results_dict, dict):
            raise ValueError("GloWbe results must be a dict with genre keys")
        return results_dict

    def _extract_contexts(self, results_dict: Dict[str, Any], max_contexts: Optional[int], strategy: str) -> List[str]:
        """Extract ALL context strings from GloWbe results - no sampling, user controls input."""
        contexts = []
        
        # Handle GloWbe JSON structure: {genre_year: {filename_id: text_string}}
        for genre_year_key, filename_dict in results_dict.items():
            if not isinstance(filename_dict, dict):
                continue
            
            # Split genre_year for labeling
            parts = genre_year_key.split('_')
            genre = parts[0] if len(parts) >= 1 else 'unknown'
            year = parts[1] if len(parts) >= 2 else 'unknown'
            
            # Extract ALL text content from filename_id -> text_string mappings
            for filename_id, text_content in filename_dict.items():
                if isinstance(text_content, str) and text_content.strip():
                    # Format: [genre:year:filename_id] text_content
                    context_label = f"[{genre}:{year}:{filename_id}]"
                    contexts.append(f"{context_label} {text_content.strip()}")
                        
        return contexts

    def _extract_context_from_row(self, row, strategy: str) -> str:
        """Extract context from a single GloWbe result row."""
        text_fields = ['context', 'text', 'kwic', 'content', 'snippet']
        
        if strategy == 'first':
            for field in text_fields:
                if hasattr(row, field) and isinstance(getattr(row, field), str):
                    return getattr(row, field).strip()
        elif strategy == 'all':
            parts = []
            for field in text_fields:
                if hasattr(row, field) and isinstance(getattr(row, field), str):
                    parts.append(getattr(row, field).strip())
            return ' | '.join(parts) if parts else ''
        
        return str(row) if strategy == 'raw_json' else ''

    def _extract_context_from_item(self, item, strategy: str) -> str:
        """Extract context from a dict/object item."""
        if isinstance(item, str):
            return item
        elif isinstance(item, dict):
            text_fields = ['context', 'text', 'kwic', 'content', 'snippet']
            if strategy == 'first':
                for field in text_fields:
                    if field in item and isinstance(item[field], str):
                        return item[field].strip()
            elif strategy == 'all':
                parts = []
                for field in text_fields:
                    if field in item and isinstance(item[field], str):
                        parts.append(item[field].strip())
                return ' | '.join(parts) if parts else ''
            elif strategy == 'raw_json':
                return json.dumps(item)
        return str(item)

    def _compute_GloWbe_stats(self, results_dict: Dict[str, Any], keyword: str, strategy: str) -> Dict[str, Any]:
        """Compute statistics about GloWbe results distribution - handles {genre_year: {filename_id: text_string}} structure."""
        # Extract genres from genre_year keys
        genres = set()
        genre_counts = {}
        total_contexts = 0
        
        for genre_year_key, filename_dict in results_dict.items():
            if isinstance(filename_dict, dict):
                # Split to get genre
                parts = genre_year_key.split('_')
                if len(parts) >= 1:
                    genre = parts[0]
                    genres.add(genre)
                    
                    # Count contexts (filename entries)
                    context_count = len(filename_dict)
                    total_contexts += context_count
                    
                    # Aggregate by genre
                    if genre not in genre_counts:
                        genre_counts[genre] = 0
                    genre_counts[genre] += context_count
        
        return {
            'keyword': keyword,
            'genres': sorted(list(genres)),
            'genre_counts': genre_counts,
            'total_contexts': total_contexts,
            'extraction_strategy': strategy
        }

    def _build_GloWbe_prompt(self, keyword: str, results_dict: Dict[str, Any], stats: Dict[str, Any], 
                          analysis_focus: str, max_contexts: Optional[int], return_json: bool, 
                          extraction_strategy: str) -> str:
        """Build specialized prompt for GloWbe forensic linguistics analysis."""
        
        contexts = self._extract_contexts(results_dict, max_contexts, extraction_strategy)
        
        # Build genre summary from actual stats, not just what's in contexts
        genre_summary = ", ".join([f"{g}({stats['genre_counts'][g]})" for g in stats['genres']])
        
        # Add explicit instruction about complete data inclusion
        contexts_section = f"""GloWbe KWIC Contexts (ALL {len(contexts)} contexts from provided data):
---
IMPORTANT: ALL contexts from your provided GloWbe data are included below: {genre_summary}
Each context is labeled [genre:year:filename_id] to show its source.
No sampling or filtering was performed - this is your complete dataset.
---
""" + "\n".join(contexts) + "\n---\n"
        
        focus_instructions = {
            "forensic_linguistics": """
            As a computational forensic linguist, perform systematic analysis to identify:
            1. **Semantic Range Mapping**: Document all distinct senses/meanings of the keyword
            2. **Register Variation**: Compare usage patterns across genres (academic, news, fiction, etc.)
            3. **Collocational Profiles**: Identify key collocates and their significance
            4. **Frequency Distributions**: Analyze genre-specific frequency patterns
            5. **Interpretive Stability**: Assess semantic consistency vs. context-dependency
            6. **Forensic Implications**: Note patterns relevant to authorship, text dating, or authenticity
            """,
            "semantic_variation": """
            Focus on semantic analysis:
            1. Identify polysemy patterns and meaning boundaries
            2. Map semantic fields and conceptual domains
            3. Analyze metaphorical vs. literal usage
            4. Document semantic change indicators across contexts
            """,
            "register_analysis": """
            Perform register-specific analysis:
            1. Compare formal vs. informal usage patterns
            2. Identify genre-specific conventions
            3. Analyze technical vs. general usage
            4. Map sociolinguistic variation patterns
            """,
            "diachronic": """
            Analyze temporal patterns:
            1. Identify usage evolution across time periods
            2. Map emerging vs. declining meanings
            3. Track semantic change trajectories
            4. Document historical usage patterns
            """,
            "comparative": """
            Perform comparative analysis:
            1. Cross-genre pattern comparison
            2. Usage frequency analysis
            3. Contextual distribution mapping
            4. Identify genre-specific markers
            """
        }
        
        base_prompt = f"""
        You are a computational forensic linguistics AI agent analyzing GloWbe (Contemporary Corpus of American English) data.

        METHODOLOGICAL FRAMEWORK:
        Apply systematic data science, legal scholarship, and applied linguistics approaches to analyze the keyword "{keyword}".

        CORPUS DATA SUMMARY:
        - Keyword: "{keyword}"
        - Total Contexts Provided: {stats['total_contexts']:,} across {len(results_dict)} genre_year combinations
        - Genre Distribution: {genre_summary}
        - Contexts Analyzed: ALL {len(contexts)} contexts (complete dataset, no sampling)
        - Extraction Strategy: {extraction_strategy}
        
        ANALYSIS FOCUS: {analysis_focus}
        {focus_instructions.get(analysis_focus, focus_instructions['forensic_linguistics'])}

        SYSTEMATIC STEPS:
        1. **Data Overview**: Summarize distribution across ALL genres (use the counts provided above)
        2. **Pattern Recognition**: Identify recurring usage patterns across different genres
        3. **Statistical Analysis**: Note frequency and distribution patterns across ALL genres
        4. **Linguistic Analysis**: Analyze syntactic, semantic, and pragmatic features by genre
        5. **Forensic Assessment**: Evaluate evidential value for text analysis
        6. **Interpretive Framework**: Provide systematic interpretation guidelines

        CRITICAL CONSTRAINTS:
        - Use ALL the provided GloWbe contexts (complete dataset as provided by user)
        - Apply rigorous linguistic methodology across all provided contexts
        - Avoid speculation beyond evidence
        - Maintain scientific objectivity
        - Analyze the complete distribution of contexts as provided (no sampling performed)

        {contexts_section}
        """
        
        if return_json:
            base_prompt += """
            Return ONLY valid JSON with this structure:
            {
              "keyword": string,
              "total_contexts": number,
              "genre_distribution": object,
              "reasoning_content": [string, ...],
              "semantic_analysis": string,
              "register_patterns": string,
              "forensic_implications": string,
              "summary": string,
              "limitations": string
            }
            """
        else:
            base_prompt += """
            Provide structured analysis with these sections:
            1. **Corpus Distribution Overview** (use the full genre counts provided)
            2. **Semantic Analysis** 
            3. **Register and Genre Patterns** (analyze patterns across ALL genres)
            4. **Collocational Analysis**
            5. **Forensic Linguistics Assessment**
            6. **Interpretive Guidelines**
            7. **Methodological Limitations**
            """
        
        return base_prompt.strip()

    def _postprocess_GloWbe_json(self, content: str, stats: Dict[str, Any]) -> Dict[str, Any]:
        """Process and validate JSON response from model."""
        try:
            parsed = json.loads(content)
        except Exception:
            # Try to extract JSON from response
            match = re.search(r'{[\s\S]*}', content)
            if match:
                try:
                    parsed = json.loads(match.group(0))
                except Exception:
                    parsed = None
            else:
                parsed = None
        
        if not isinstance(parsed, dict):
            # Fallback structure
            parsed = {
                "keyword": stats['keyword'],
                "total_contexts": stats['total_contexts'],
                "genre_distribution": stats['genre_counts'],
                "reasoning_content": [
                    "Model did not return valid JSON; content auto-wrapped.",
                    "Analysis limited by response format issues."
                ],
                "semantic_analysis": content if isinstance(content, str) else str(content),
                "register_patterns": "Unable to extract due to format issues.",
                "forensic_implications": "Analysis inconclusive due to response parsing failure.",
                "summary": "Response required manual wrapping - review raw content.",
                "limitations": "Auto-wrapped due to invalid JSON from model."
            }
        
        # Ensure required fields exist
        required_fields = {
            "reasoning_content": [],
            "semantic_analysis": "",
            "register_patterns": "",
            "forensic_implications": "",
            "summary": "",
            "limitations": ""
        }
        
        for field, default in required_fields.items():
            if field not in parsed:
                parsed[field] = default
        
        return parsed
    
# Markdown export function for GloWbe analysis
def export_GloWbe_markdown(result, keyword: str, filename: str = None):
    """Export GloWbe forensic linguistics analysis to markdown with reasoning first."""
    import json
    from datetime import datetime
    
    def _sanitize(name: str) -> str:
        return ''.join(c if (c.isalnum() or c in ('-','_')) else '_' for c in name.strip()) or 'analysis'
    
    safe_keyword = _sanitize(keyword)
    outname = filename or f"GloWbe_forensic_{safe_keyword}.md"
    
    lines = [f"# GloWbe Forensic Linguistics Analysis: {keyword}\n\n"]
    lines.append(f"*Generated: {datetime.utcnow().isoformat()}Z*\n\n")
    
    if isinstance(result, dict):
        # Extract reasoning content first
        reasoning = result.get('reasoning_content', [])
        if reasoning:
            lines.append("## Methodological Framework\n\n")
            lines.append("```text\n")
            if isinstance(reasoning, list):
                lines.append('\n'.join(str(r) for r in reasoning))
            else:
                lines.append(str(reasoning))
            lines.append("\n```\n\n")
        
        # Add structured sections
        sections = [
            ('semantic_analysis', 'Semantic Analysis'),
            ('register_patterns', 'Register and Genre Patterns'),
            ('forensic_implications', 'Forensic Linguistics Assessment'),
            ('summary', 'Summary'),
            ('limitations', 'Limitations')
        ]
        
        for field, title in sections:
            if field in result and result[field]:
                lines.append(f"## {title}\n\n")
                lines.append(f"{result[field]}\n\n")
        
        # Add distribution data if available
        if 'genre_distribution' in result:
            lines.append("## Corpus Distribution\n\n")
            lines.append("```json\n")
            lines.append(json.dumps(result['genre_distribution'], indent=2))
            lines.append("\n```\n\n")
    
    else:
        lines.append("## Analysis\n\n")
        lines.append(str(result))
    
    content = ''.join(lines)
    
    with open(outname, 'w', encoding='utf-8') as f:
        f.write(content)
    
    print(f"üìÑ GloWbe forensic analysis exported: {outname} ({len(content)} chars)")
    return outname

print("‚úÖ GloWbe Forensic Linguistics Tool loaded!")
````

````python
# Minimal Coding Classification Tool
# Simplified single-category classification: literal, metaphorical, neither, or unclear

from langchain.tools import BaseTool
from langchain.pydantic_v1 import BaseModel, Field
from langchain.chat_models import init_chat_model
from typing import Optional, Type, Dict, Any, Union, List
import json
import pandas as pd
from datetime import datetime
import os
from pathlib import Path


class MinimalCodingInput(BaseModel):
    """Input for minimal coding classification."""
    keyword: str = Field(description="The keyword being analyzed (e.g., 'best system')")
    results_json: Union[str, Dict[str, Any]] = Field(
        description="KWIC JSON results from got3.search_keyword_corpus"
    )
    batch_size: Optional[int] = Field(
        default=100, 
        description="Number of concordance lines to process per batch"
    )
    reasoning_level: Optional[str] = Field(
        default="medium", 
        description="Level of reasoning detail: 'high', 'medium', 'low'"
    )
    return_dataframe: bool = Field(
        default=True, 
        description="If True, return pandas DataFrame; if False, return list of dicts"
    )
    cache_dir: Optional[str] = Field(
        default="./cache/minimal_coding_classifications",
        description="Directory to cache batch results (prevents loss on failure)"
    )
    use_cache: bool = Field(
        default=True,
        description="If True, use cached results and resume from last completed batch"
    )
    debug: bool = Field(default=False, description="Enable debug output")


class MinimalCodingClassificationTool(BaseTool):
    """
    AI tool for minimal coding classification of concordance lines.
    
    Classifies each concordance line into ONE category:
    - literal: The keyword is used in its literal/direct sense
    - figurative: The keyword is used figuratively or in an indirect sense
    - neither: The usage doesn't clearly fit either category
    - unclear: Insufficient context to determine
    
    Features resilient caching to prevent loss of expensive API calls on failure.
    """
    name: str = "minimal_coding_classification"
    description: str = (
        "Classifies concordance lines into a single category: "
        "literal, figurative, neither, or unclear"
    )
    args_schema: Type[BaseModel] = MinimalCodingInput
    model: Any = Field(exclude=True)

    def __init__(self, model, **kwargs):
        super().__init__(**kwargs)
        self.model = model

    def _run(
        self,
        keyword: str,
        results_json: Union[str, Dict[str, Any]],
        reasoning_level: str = "medium",
        batch_size: int = 100,
        return_dataframe: bool = True,
        cache_dir: str = "./cache/minimal_coding_classifications",
        use_cache: bool = True,
        debug: bool = False,
    ) -> Union[pd.DataFrame, List[Dict[str, Any]], str]:
        try:
            return self._execute(keyword, results_json, reasoning_level,
                               batch_size, return_dataframe, cache_dir, use_cache, debug)
        except Exception as e:
            import traceback
            error_details = traceback.format_exc()
            print(f"‚ùå Error during minimal coding classification:")
            print(error_details)
            return f"‚ùå Error: {str(e)}\n\n{error_details}"

    async def _arun(
        self,
        keyword: str,
        results_json: Union[str, Dict[str, Any]],
        reasoning_level: str = "medium",
        batch_size: int = 100,
        return_dataframe: bool = True,
        cache_dir: str = "./cache/minimal_coding_classifications",
        use_cache: bool = True,
        debug: bool = False,
    ) -> Union[pd.DataFrame, List[Dict[str, Any]], str]:
        return self._run(keyword, results_json, reasoning_level,
                          batch_size, return_dataframe, cache_dir, use_cache, debug)

    def _execute(self, keyword, results_json, reasoning_level,
                batch_size, return_dataframe, cache_dir, use_cache, debug):
        # Parse input
        results_dict = self._parse_results(results_json)
        
        # Extract all concordance lines with metadata
        concordance_lines = self._extract_concordance_lines(results_dict, keyword)
        
        if debug:
            print(f"‚úÖ Extracted {len(concordance_lines)} concordance lines for '{keyword}'")
            print(f"üìä Processing in batches of {batch_size}")
        
        # Validate reasoning level
        if reasoning_level not in ['high', 'medium', 'low']:
            reasoning_level = 'medium'
        
        # Setup cache directory
        cache_path = Path(cache_dir)
        if use_cache:
            cache_path.mkdir(parents=True, exist_ok=True)
            session_id = datetime.now().strftime("%Y%m%d_%H%M%S")
            cache_prefix = f"{self._sanitize_filename(keyword)}_{reasoning_level}_b{batch_size}_{session_id}"
            if debug:
                print(f"üíæ Cache enabled: {cache_path / cache_prefix}_*.json")
        
        # Check for existing cache files and load them
        all_classifications = []
        completed_batches = set()
        
        if use_cache:
            safe_keyword = self._sanitize_filename(keyword)
            existing_caches = list(cache_path.glob(f"{safe_keyword}_{reasoning_level}_b{batch_size}_*_batch_*.json"))
            if existing_caches:
                print(f"üìÇ Found {len(existing_caches)} cached batch files")
                for cache_file in sorted(existing_caches):
                    try:
                        with open(cache_file, 'r') as f:
                            cached_batch = json.load(f)
                            all_classifications.extend(cached_batch['classifications'])
                            completed_batches.add(cached_batch['batch_num'])
                        if debug:
                            print(f"  ‚úÖ Loaded batch {cached_batch['batch_num']} from cache")
                    except Exception as e:
                        print(f"  ‚ö†Ô∏è Failed to load {cache_file.name}: {e}")
                
                if completed_batches:
                    print(f"‚ôªÔ∏è Resuming from batch {max(completed_batches) + 1} (loaded {len(all_classifications)} classifications)")
        
        # Process in batches with progress bar
        from tqdm import tqdm as std_tqdm
        
        total_batches = (len(concordance_lines) + batch_size - 1) // batch_size
        
        print(f"\nüîÑ Classifying {len(concordance_lines)} concordance lines in {total_batches} batches...")
        
        for batch_idx in range(0, len(concordance_lines), batch_size):
            batch_num = (batch_idx // batch_size) + 1
            
            # Skip if this batch was already completed
            if batch_num in completed_batches:
                continue
            
            batch = concordance_lines[batch_idx:batch_idx + batch_size]
            
            # Print progress
            print(f"üìù Batch {batch_num}/{total_batches} ({len(batch)} lines)... ", end='', flush=True)
            
            # Classify this batch
            batch_results = self._classify_batch(keyword, batch, reasoning_level, debug)
            all_classifications.extend(batch_results)
            
            print(f"‚úì", flush=True)
            
            # Cache this batch immediately
            if use_cache:
                cache_filename = f"{cache_prefix}_batch_{batch_num:04d}.json"
                cache_filepath = cache_path / cache_filename
                try:
                    cache_data = {
                        'keyword': keyword,
                        'reasoning_level': reasoning_level,
                        'batch_size': batch_size,
                        'batch_num': batch_num,
                        'total_batches': total_batches,
                        'timestamp': datetime.now().isoformat(),
                        'classifications': batch_results
                    }
                    with open(cache_filepath, 'w') as f:
                        json.dump(cache_data, f, indent=2)
                except Exception as e:
                    if debug:
                        print(f"  ‚ö†Ô∏è Failed to cache batch {batch_num}: {e}")
        
        if debug:
            print(f"\n‚úÖ Classification complete: {len(all_classifications)} lines classified")
        
        # Return as DataFrame or list
        if return_dataframe:
            return pd.DataFrame(all_classifications)
        return all_classifications

    def _sanitize_filename(self, name: str) -> str:
        """Sanitize filename to remove problematic characters."""
        return ''.join(c if (c.isalnum() or c in ('-','_')) else '_' for c in name.strip()) or 'keyword'

    def _parse_results(self, results_json: Union[str, Dict[str, Any]]) -> Dict[str, Any]:
        """Parse results JSON."""
        if isinstance(results_json, str):
            return json.loads(results_json)
        return results_json

    def _extract_concordance_lines(self, results_dict: Dict[str, Any], keyword: str) -> List[Dict[str, Any]]:
        """
        Extract all concordance lines with metadata from results.
        Structure: {genre_year: {filename_id: text_string}}
        """
        concordance_lines = []
        line_id = 1
        
        for genre_year_key, filename_dict in results_dict.items():
            if not isinstance(filename_dict, dict):
                continue
            
            # Parse genre and year from key
            parts = genre_year_key.split('_')
            genre = parts[0] if len(parts) >= 1 else 'unknown'
            year = parts[1] if len(parts) >= 2 else 'unknown'
            
            for filename_id, text_content in filename_dict.items():
                if isinstance(text_content, str) and text_content.strip():
                    # Sanitize text
                    sanitized_text = self._sanitize_text(text_content)
                    
                    # Skip if text is too short or too long
                    if len(sanitized_text) < 10 or len(sanitized_text) > 5000:
                        continue
                    
                    concordance_lines.append({
                        'line_id': line_id,
                        'genre': genre,
                        'year': year,
                        'genre_year': genre_year_key,
                        'filename_id': filename_id,
                        'concordance_line': sanitized_text,
                        'keyword': keyword
                    })
                    line_id += 1
        
        return concordance_lines
    
    def _sanitize_text(self, text: str) -> str:
        """Sanitize text to remove problematic characters and normalize whitespace."""
        import re
        
        # Remove control characters except newlines and tabs
        text = ''.join(char for char in text if ord(char) >= 32 or char in '\n\t')
        
        # Normalize whitespace
        text = re.sub(r'\s+', ' ', text)
        
        # Remove NULL bytes
        text = text.replace('\x00', '')
        
        return text.strip()

    def _classify_batch(self, keyword: str, batch: List[Dict[str, Any]], 
                       reasoning_level: str, debug: bool) -> List[Dict[str, Any]]:
        """Classify a batch of concordance lines using the AI model."""
        
        # Build prompt for batch classification
        prompt = self._build_classification_prompt(keyword, batch, reasoning_level)
        
        # Invoke model with timeout protection
        import signal
        from contextlib import contextmanager
        
        @contextmanager
        def timeout(seconds):
            """Context manager for timeout handling."""
            def timeout_handler(signum, frame):
                raise TimeoutError(f"Model invocation timed out after {seconds} seconds")
            
            old_handler = signal.signal(signal.SIGALRM, timeout_handler)
            signal.alarm(seconds)
            try:
                yield
            finally:
                signal.alarm(0)
                signal.signal(signal.SIGALRM, old_handler)
        
        try:
            # Set timeout to 120 seconds per batch
            with timeout(120):
                response = self.model.invoke([{"role": "user", "content": prompt}])
        except TimeoutError as e:
            if debug:
                print(f"  ‚ö†Ô∏è Timeout: {e}")
            # Return batch with timeout error
            return self._create_error_batch(batch, 'Timeout after 120 seconds')
        except Exception as e:
            if debug:
                print(f"  ‚ö†Ô∏è Model invocation error: {e}")
            return self._create_error_batch(batch, f'Model error: {str(e)}')
        
        # Extract content from response
        content = response.content if hasattr(response, 'content') else str(response)
        
        # Parse JSON response
        try:
            classifications = self._parse_classification_response(content, batch)
        except Exception as e:
            if debug:
                print(f"‚ö†Ô∏è Failed to parse batch classification: {e}")
                print(f"Raw response (first 500 chars): {str(content)[:500]}")
            return self._create_error_batch(batch, f'Failed to parse model response: {str(e)}')
        
        return classifications

    def _create_error_batch(self, batch: List[Dict[str, Any]], error_msg: str) -> List[Dict[str, Any]]:
        """Create error classifications for a batch."""
        classifications = []
        for line in batch:
            classification = {
                'line_id': line['line_id'],
                'genre': line['genre'],
                'year': line['year'],
                'genre_year': line['genre_year'],
                'filename_id': line['filename_id'],
                'concordance_line': line['concordance_line'],
                'keyword': line['keyword'],
                'classification': 'unclear',
                'reasoning': error_msg
            }
            classifications.append(classification)
        return classifications

    def _build_classification_prompt(self, keyword: str, batch: List[Dict[str, Any]], 
                                    reasoning_level: str) -> str:
        """Build the classification prompt for minimal coding approach."""
        
        # Build concordance lines section
        lines_text = []
        for line in batch:
            lines_text.append(f"Line {line['line_id']}: {line['concordance_line']}")
        
        prompt = f"""

# BACKGROUND INFORMATION
System Prompt: You are a chatbot assisting legal scholars and linguists by analyzing concordance lines
Reasoning Level: High 
Court Case: West Virginia v. Environmental Protection Agency, 597 U.S. ___ (2022)
Court Case Summary: In 2015, the Environmental Protection Agency (EPA) promulgated the Clean Power Plan rule, which addressed carbon dioxide emissions from existing power plants, citing Section 111 of the Clean Air Act," 42 U.S.C. 7411(d). Although the states set the enforceable rules governing existing sources, EPA determines the emissions limit with which they have to comply by determining the "best system of emission reduction" (BSER). In the Clean Power Plan, EPA determined that the BSER for existing coal and natural gas plants included "heat rate improvements" at coal-fired plants and "generation-shifting," i.e., a shift in electricity production from existing coal-fired to natural-gas-fired plants and from both coal and gas plants to renewables (wind and solar). An operator could reduce the regulated plant's production of electricity, build or invest in new or existing equipment, or purchase emission allowances as part of a cap-and-trade regime. No existing coal plant could achieve the emissions performance rates without generation-shifting. The Supreme Court stayed the Clean Power Plan in 2016. It was later repealed when EPA determined that it lacked authority "of this breadth." EPA then promulgated the Affordable Clean Energy (ACE) rule, mandating equipment upgrades and operating practices. The D.C. Circuit held that EPA's repeal of the Clean Power Plan rested on a mistaken reading of the Clean Air Act and vacated the ACE rule. The Supreme Court reversed. Congress did not grant EPA the authority to devise emissions caps based on the Clean Power Plan's generation-shifting approach. Restructuring the nation's mix of electricity generation cannot be the BSER under Section 111. Under the major questions doctrine, an agency must point to "clear congressional authorization" for such an unprecedented exercise of authority. On EPA's view of Section 111(d), Congress implicitly tasked it alone with balancing vital considerations of national policy. Issues of electricity transmission and distribution are not within EPA's traditional expertise. The Clean Power Plan "conveniently enabled" EPA to enact a program, cap-and-trade, that Congress rejected numerous times.

# TASK
Classify each concordance line for the keyword "best system" into ONE of these categories, to determine if the ordinary meaning aligns with the notion that "Restructuring the nation's mix of electricity generation cannot be the BSER under Section 111"

1. **literal** - The keyword is used directly, in a literal sense, to refer to a specific system or method that is objectively the ideal or optimal choice
2. **figurative** - The keyword is used indirectly, in a figurative sense, to refer to a system or method that may be considered the best but is known to be subjective, context-dependent, or sub-optimal
3. **neither** - The usage doesn't clearly fit literal or figurative categories
4. **unclear** - Insufficient context to determine the usage type

# CRITICAL INSTRUCTIONS
1. **Text-Internal Evidence Only**: Base your classification SOLELY on evidence in the concordance line itself
2. **Single Category**: Each line gets EXACTLY ONE classification
3. **Brief Reasoning**: Provide 1-2 sentences explaining your classification based on text-internal evidence
4. **Conservative Approach**: When in doubt, use "unclear"

# CONCORDANCE LINES TO CLASSIFY
{chr(10).join(lines_text)}

# OUTPUT FORMAT
Return ONLY a valid JSON array with one object per line:
[
  {{
    "line_id": 1,
    "classification": "literal|figurative|neither|unclear",
    "reasoning": "Brief explanation based on text-internal evidence"
  }},
  ...
]

Provide NO other text - only the JSON array.
"""
        
        return prompt.strip()

    def _parse_classification_response(self, content: Any, batch: List[Dict[str, Any]]) -> List[Dict[str, Any]]:
        """Parse the model's JSON response and merge with original metadata."""
        
        # Handle case where content is a list (Bedrock response structure)
        if isinstance(content, list):
            text_parts = []
            for item in content:
                if isinstance(item, dict):
                    if 'text' in item:
                        text_parts.append(item['text'])
                    elif 'content' in item:
                        text_parts.append(item['content'])
                elif isinstance(item, str):
                    text_parts.append(item)
            content = ''.join(text_parts)
        
        # Convert to string and strip
        content = str(content).strip()
        
        # Remove markdown code blocks if present
        if content.startswith('```'):
            lines = content.split('\n')
            content = '\n'.join(lines[1:-1] if len(lines) > 2 else lines)
            content = content.replace('```json', '').replace('```', '').strip()
        
        # Parse JSON
        classifications_raw = json.loads(content)
        
        if not isinstance(classifications_raw, list):
            raise ValueError("Expected JSON array of classifications")
        
        # Merge classifications with original metadata
        results = []
        classification_map = {c.get('line_id'): c for c in classifications_raw}
        
        for line in batch:
            line_id = line['line_id']
            classification = {
                'line_id': line_id,
                'genre': line['genre'],
                'year': line['year'],
                'genre_year': line['genre_year'],
                'filename_id': line['filename_id'],
                'concordance_line': line['concordance_line'],
                'keyword': line['keyword']
            }
            
            # Add classification if available
            if line_id in classification_map:
                classified = classification_map[line_id]
                classification['classification'] = classified.get('classification', 'unclear')
                classification['reasoning'] = classified.get('reasoning', '')
            else:
                # Missing classification
                classification['classification'] = 'unclear'
                classification['reasoning'] = 'No classification returned by model'
            
            results.append(classification)
        
        return results


def export_minimal_coding_to_excel(df: pd.DataFrame, keyword: str, filename: str = None):
    """Export minimal coding classifications to Excel."""
    
    def _sanitize(name: str) -> str:
        return ''.join(c if (c.isalnum() or c in ('-','_')) else '_' for c in name.strip()) or 'classifications'
    
    safe_keyword = _sanitize(keyword)
    outname = filename or f"minimal_coding_{safe_keyword}_{datetime.now().strftime('%Y%m%d_%H%M%S')}.xlsx"
    
    # Reorder columns for readability
    column_order = ['line_id', 'keyword', 'classification', 'reasoning', 'concordance_line', 
                    'genre', 'year', 'genre_year', 'filename_id']
    
    # Only include columns that exist
    export_cols = [col for col in column_order if col in df.columns]
    df_export = df[export_cols].copy()
    
    # Export to Excel
    df_export.to_excel(outname, index=False, sheet_name='Classifications')
    
    print(f"üìä Classifications exported to Excel: {outname}")
    print(f"   Total lines: {len(df_export)}")
    
    # Print summary statistics
    if 'classification' in df.columns:
        print(f"\nüìà Classification Summary:")
        counts = df['classification'].value_counts()
        for value, count in counts.items():
            pct = (count / len(df) * 100)
            print(f"   {value}: {count} ({pct:.1f}%)")
    
    return outname


print("‚úÖ Minimal Coding Classification Tool loaded!")
print("üìã Categories: literal, figurative, neither, unclear")
print("üíæ Caching enabled by default to prevent loss on failure")
````

## Corpus Pattern Analysis: Pre and Post Modifiers

Extract linguistic patterns around "best system" to analyze:
- Pre-modifiers (words before "best system")
- Post-modifiers (words after "best system") 
- Part-of-speech patterns
- Dependency relationships

````python
df1.concordance_line = df1.concordance_line.apply(lambda x: x.replace('**best','best'))
df1.concordance_line = df1.concordance_line.apply(lambda x: x.replace('system','system**'))
````

````python
# Corpus Pattern Analysis: Extract pre and post modifiers for "best system"

import pandas as pd
import spacy

# Load spaCy model if not already loaded
if 'nlp' not in locals():
    nlp = spacy.load("en_core_web_lg")

# Function to find "best system" and extract modifiers
def extract_modifiers(text, keyword_phrase="best system", window_size=5):
    """
    Extract pre and post modifiers around a keyword phrase.
    
    Args:
        text: The concordance line text
        keyword_phrase: The phrase to search for (default: "best system")
        window_size: Number of tokens to extract before/after (default: 5)
    
    Returns:
        dict with pre_modifiers, post_modifiers, and other linguistic features
    """
    doc = nlp(text)
    keyword_tokens = keyword_phrase.lower().split()
    
    # Find the position of "best system" in the text
    for i in range(len(doc) - len(keyword_tokens) + 1):
        # Check if we found the keyword phrase
        match = all(
            doc[i + j].text.lower() == keyword_tokens[j] 
            for j in range(len(keyword_tokens))
        )
        
        if match:
            # Found the keyword phrase at position i
            keyword_start = i
            keyword_end = i + len(keyword_tokens)
            
            # Extract pre-modifiers (words before keyword)
            pre_start = max(0, keyword_start - window_size)
            pre_tokens = doc[pre_start:keyword_start]
            
            # Extract post-modifiers (words after keyword)
            post_end = min(len(doc), keyword_end + window_size)
            post_tokens = doc[keyword_end:post_end]
            
            # Get the keyword span
            keyword_span = doc[keyword_start:keyword_end]
            
            # Extract linguistic features
            result = {
                'keyword_found': True,
                'keyword_index': keyword_start,
                'keyword_text': keyword_span.text,
                
                # Pre-modifiers
                'pre_tokens': [t.text for t in pre_tokens],
                'pre_lemmas': [t.lemma_ for t in pre_tokens],
                'pre_pos': [t.pos_ for t in pre_tokens],
                'pre_tags': [t.tag_ for t in pre_tokens],
                'pre_text': pre_tokens.text if pre_tokens else '',
                
                # Post-modifiers
                'post_tokens': [t.text for t in post_tokens],
                'post_lemmas': [t.lemma_ for t in post_tokens],
                'post_pos': [t.pos_ for t in post_tokens],
                'post_tags': [t.tag_ for t in post_tokens],
                'post_text': post_tokens.text if post_tokens else '',
                
                # Immediate neighbors (1 word before/after)
                'immediate_pre': doc[keyword_start - 1].text if keyword_start > 0 else '',
                'immediate_post': doc[keyword_end].text if keyword_end < len(doc) else '',
                'immediate_pre_pos': doc[keyword_start - 1].pos_ if keyword_start > 0 else '',
                'immediate_post_pos': doc[keyword_end].pos_ if keyword_end < len(doc) else '',
                
                # Dependency info for keyword tokens
                'keyword_deps': [t.dep_ for t in keyword_span],
                'keyword_heads': [t.head.text for t in keyword_span],
                
                # Full sentence
                'full_text': text
            }
            
            return result
    
    # Keyword not found
    return {
        'keyword_found': False,
        'full_text': text
    }

# Test on a sample
sample_text = "The more complex an issue is, the less possible it is to actually know the best system in advance"
sample_result = extract_modifiers(sample_text)

print("üìä Sample extraction:")
print(f"Pre-modifiers: {sample_result.get('pre_text', 'N/A')}")
print(f"Keyword: {sample_result.get('keyword_text', 'N/A')}")
print(f"Post-modifiers: {sample_result.get('post_text', 'N/A')}")
print(f"\nImmediate context: '{sample_result.get('immediate_pre', '')}' | best system | '{sample_result.get('immediate_post', '')}'")
print(f"\nPOS tags (pre): {sample_result.get('pre_pos', [])}")
print(f"POS tags (post): {sample_result.get('post_pos', [])}")
````

````python
# Apply pattern extraction to your DataFrame (df1 or df0)
# Assuming df1 has a 'concordance_line' column with the text

# Choose which DataFrame to analyze
df = df1  # or df0, depending on which you want to analyze

print(f"üìä Processing {len(df)} concordance lines...")

# Extract modifiers for each row
pattern_data = []
for idx, row in df.iterrows():
    # Get the concordance text (adjust column name as needed)
    text = row.get('concordance_line', row.get('text', row.get('context', '')))
    
    # Extract modifiers
    result = extract_modifiers(text, keyword_phrase="best system", window_size=5)
    
    # Add row metadata
    result['line_id'] = row.get('line_id', idx)
    result['genre'] = row.get('genre', '')
    result['year'] = row.get('year', '')
    result['classification'] = row.get('classification', '')
    
    pattern_data.append(result)

# Create new DataFrame with pattern analysis
pattern_df = pd.DataFrame(pattern_data)

# Filter to only rows where keyword was found
pattern_df_found = pattern_df[pattern_df['keyword_found'] == True].copy()

print(f"‚úÖ Extraction complete!")
print(f"   Total lines processed: {len(pattern_df)}")
print(f"   Keyword found in: {len(pattern_df_found)} lines")
print(f"\nüìã Pattern DataFrame columns:")
print(f"   {list(pattern_df_found.columns)}")

# Display sample
pattern_df_found.head()
````

## AI-Assisted Corpus Pattern Analysis

Use AI to analyze POS patterns and identify semantic differences between **literal** and **figurative** classifications of "best system"

````python
# AI-Assisted Corpus Pattern Analysis Function
from langchain.chat_models import init_chat_model
import json
from datetime import datetime

def ai_corpus_pattern_analysis(pos_json_data, classification_filter=None, 
                                model_id='openai.gpt-oss-120b-1:0',
                                temperature=0.0):
    """
    Use AI to analyze POS patterns and semantic differences across classifications.
    
    Args:
        pos_json_data: JSON string or dict with pattern analysis data
        classification_filter: Filter by classification ('literal', 'figurative', etc.) or None for all
        model_id: AWS Bedrock model to use
        temperature: Model temperature (0.0 for deterministic)
    
    Returns:
        AI analysis of corpus patterns
    """
    
    # Parse JSON if string
    if isinstance(pos_json_data, str):
        data = json.loads(pos_json_data)
    else:
        data = pos_json_data
    
    # Convert to DataFrame for easier filtering
    df = pd.DataFrame(data)
    
    # Filter by classification if specified
    if classification_filter:
        df = df[df['classification'] == classification_filter]
        filter_desc = f" (filtered to classification='{classification_filter}')"
    else:
        filter_desc = " (all classifications)"
    
    print(f"üìä Analyzing {len(df)} concordance lines{filter_desc}")
    
    # Group by classification for comparison
    classifications = df['classification'].value_counts().to_dict()
    
    # Sample data for each classification (to keep within token limits)
    sample_size = 10
    samples_by_class = {}
    for classification, group in df.groupby('classification'):
        sample = group.head(sample_size)
        samples_by_class[classification] = sample[['immediate_pre', 'immediate_post', 
                                                     'immediate_pre_pos', 'immediate_post_pos',
                                                     'pre_tokens', 'post_tokens', 
                                                     'pre_pos', 'post_pos',
                                                     'full_text']].to_dict('records')
    
    # Build AI prompt
    prompt = f"""
You are a computational corpus linguist analyzing POS (Part-of-Speech) patterns for the phrase "best system" 
in the context of legal linguistics and statutory interpretation.

# BACKGROUND
This analysis comes from concordance lines of "best system" from the COCA corpus, where each line has been 
classified by AI as:
- **literal**: Direct, literal meaning (genuinely optimal, perfect system)
- **figurative**: Indirect, figurative meaning (comparatively best, "best we have", sarcastic)
- **neither**: Doesn't fit either category
- **unclear**: Insufficient context

The legal question: Does the ordinary meaning of "best system" in the Clean Air Act support a broad or narrow 
reading of EPA's authority to restructure the electricity generation mix?

# CORPUS DATA SUMMARY
Total lines analyzed: {len(df)}
Classification distribution:
{json.dumps(classifications, indent=2)}

# PATTERN DATA BY CLASSIFICATION
Below are sample concordance lines with their POS patterns, grouped by classification:

{json.dumps(samples_by_class, indent=2)}

# ANALYSIS TASK
Perform a systematic Corpus Pattern Analysis to answer:

1. **Pre-modifier Patterns**: Are there distinct POS or lexical patterns BEFORE "best system" that distinguish 
   literal from figurative usage? (e.g., determiners, adjectives, modals)

2. **Post-modifier Patterns**: Are there distinct POS or lexical patterns AFTER "best system" that distinguish 
   literal from figurative usage? (e.g., prepositional phrases, relative clauses)

3. **Semantic Collocations**: What words frequently co-occur with "best system" in literal vs figurative contexts?
   Do these reveal semantic differences?

4. **Syntactic Structures**: Are there syntactic constructions more common in literal vs figurative usage?
   (e.g., "the best system" vs "a best system", "best system for X" vs "best system available")

5. **Interpretive Implications**: Based on these patterns, what can we infer about the ORDINARY MEANING of 
   "best system" as understood by competent English speakers? Does the corpus evidence support:
   - A narrow reading (literal/optimal system)
   - A broad reading (comparative/flexible system)
   - Both readings are attested in ordinary usage?

# OUTPUT FORMAT
Provide a structured analysis with:
- **Key Findings**: Bullet points of major pattern differences
- **Literal Usage Patterns**: Distinctive features of literal classifications
- **Figurative Usage Patterns**: Distinctive features of figurative classifications
- **Statistical Observations**: Frequency patterns, collocations, POS distributions
- **Legal Linguistics Implications**: What this means for statutory interpretation
- **Corpus Linguistics Conclusion**: Evidence-based assessment of ordinary meaning

Be specific, cite examples from the data, and maintain scientific rigor.
"""
    
    # Initialize model
    print(f"ü§ñ Initializing AI model: {model_id}")
    model = init_chat_model(
        model_id,
        model_provider="bedrock_converse",
        credentials_profile_name='atn-developer',
        max_tokens=128000,
        temperature=temperature
    )
    
    # Invoke model
    print(f"üî¨ Running AI corpus pattern analysis...")
    response = model.invoke([{"role": "user", "content": prompt}])
    
    # Extract content
    content = getattr(response, 'content', str(response))
    
    return content

# Test message
print("‚úÖ AI Corpus Pattern Analysis function loaded")
print("üìã Usage: ai_corpus_pattern_analysis(pos_json, classification_filter='literal')")
print("   or: ai_corpus_pattern_analysis(pos_json)  # analyze all classifications")
````
